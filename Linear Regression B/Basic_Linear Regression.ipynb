## Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_mc,drop_first = True)
```

```python
data_with_dummies.head()
```

### Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear regression model

### Declare the inputs and the targets

```python
#split the data
targets = data_preprocessed["log_price"]
inputs = data_preprocessed.drop(["log_price"],axis=1)
```

## scale the inputs

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

### Train Test Split


```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(input_scaled, targets, test_size = 0.2, random_state =365)
```

### Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#y_hat = y_pred
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel("Target(y-train)",size=18)
plt.ylabel("Predictions(y_hat)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show
```

```python
sns.displot(y_train-y_hat,kde= True)
plt.title("Residuals PDF",size =18)
```

```python
reg.score(x_train,y_train)
```

### Finding weights and Bias>>>> intercept and slop >>>> m and c


```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ["Features"])
reg_summary["Weights"]= reg.coef_
reg_summary
```

### positive weight shows : features increase in values, so do the log_price and "price respectively"
* negative weight shows : features increase in values, so do the log_price and "price decreases"

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha =0.2)
plt.xlabel("Target(y-test)",size=18)
plt.ylabel("Predictions(y_hat_test)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show
```

```python
df_pf = pd.DataFrame(y_hat_test,columns = ["Prediction"])
df_pf.head()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns = ["Prediction"])
df_pf.head()
```

```python
df_pf["Target"]= np.exp(y_test)
df_pf
```

```python
y_test
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf["Target"] = np.exp(y_test)
df_pf
```

```python
df_pf["Residual"] = df_pf["Target"] - df_pf["Prediction"]
```

```python
df_pf["Difference%"] = np.absolute(df_pf["Residual"]/df_pf["Target"]*100)
df_pf
```
